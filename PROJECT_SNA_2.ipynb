```
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.models.annotations import LabelSet
from bokeh.models.graphs import from_networkx

import pandas as pd
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```
#Membaca dataset
df_covid = pd.read_csv("crawling_covid_indonesia.csv")
```

```
df_covid = df_covid.loc[:,['username','user_mentions']]
df_covid = df_covid.dropna()
df_covid.head()
```

```
graph_covid = nx.from_pandas_edgelist(df_covid,source="username",target="user_mentions")

nx.info(graph_covid)

graph_covid.nodes()

len(graph_covid.nodes())

len(graph_covid.edges())
```

```
nx.draw
nx.draw_networkx

plt.figure(figsize=(10,10))
nx.draw(graph_covid,with_labels=True)
```

```
cliques_generator = nx.find_cliques(graph_covid)
list(cliques_generator)
```

```
help(nx.find_cliques)
```

```
all_cliques = nx.enumerate_all_cliques(graph_covid)
cliques_list = list(all_cliques)
cliques_list
```

```
k_clique_2 = [x for x in cliques_list if len(x) == 2]
```

```
print(k_clique_2)
```

```
help(nx.community.k_clique_communities)
```

```
k_comm_2 = list(nx.community.k_clique_communities(graph_covid,2))
k_comm_2
```

```
# First, let's find the 'edges' in the new graph
comm_edges = []
K = len(k_clique_2)
for u in range(K):
    for v in range(u):
        U = set(k_clique_2[u])
        V = set(k_clique_2[v])
        if len(U.intersection(V)) == 2:
            comm_edges.append((u,v))
```

```
# Next, let's build a new graph
HC = nx.Graph()
# And give it nodes...
HC.add_nodes_from(range(K))
# And the edges we calculated...
HC.add_edges_from(comm_edges)
```

```
# Finally, we use another networkx function to find the components.
comps_generator = nx.components.connected_components(HC)
```
